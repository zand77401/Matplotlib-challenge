## Observations and Insights 

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)
mouse_metadata.shape

```

```python
study_results.shape
```

```python
# Combine the data into a single dataset
merge_table = pd.merge(mouse_metadata, study_results, on="Mouse ID")

# Display the data table for preview
merge_table
```

```python
# Checking the number of mice.
merge_table['Mouse ID'].nunique()
```

```python
# Getting the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
duplicateRowsDF = merge_table[merge_table.duplicated(['Mouse ID', 'Timepoint'])]
duplicateRowsDF
```

```python
# Optional: Get all the data for the duplicate mouse ID. 

```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
clean_mouse_data = merge_table.drop_duplicates(subset=['Mouse ID', 'Timepoint'], keep='first').reset_index(drop=True)
```

```python
# Checking the number of mice in the clean DataFrame.
clean_mouse_data['Mouse ID'].nunique()
```

```python
clean_mouse_data.head(20)
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance,standard deviation, and SEM of the tumor volume for each regimen
grouped_by_drug = clean_mouse_data.groupby(['Drug Regimen'])
mean_tumor_volume = grouped_by_drug['Tumor Volume (mm3)'].mean()
median_tumor_volume = grouped_by_drug['Tumor Volume (mm3)'].median()
var_tumor_volume = grouped_by_drug['Tumor Volume (mm3)'].var()
std_tumor_volume = grouped_by_drug['Tumor Volume (mm3)'].std()
sem_tumor_volume = grouped_by_drug['Tumor Volume (mm3)'].sem()

# This method is the most straighforward, creating multiple series and putting them all together at the end.

grouped_by_drug.count().head(10)
```

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
drug_summary_table = pd.DataFrame({"Mean": mean_tumor_volume,
                                   "Median": median_tumor_volume,
                                   "Variance": var_tumor_volume,
                                   "Std Dev": std_tumor_volume,
                                   "SEM": sem_tumor_volume})
# This method produces everything in a single groupby function
print("Tumor Volume")
print("*"*80)
drug_summary_table
```
