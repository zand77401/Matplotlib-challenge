## Observations and Insights 

1) The sex of the mice were practically evenly split with 50.2% male mice and 49.8% female mice. 
2) In the box plot, the drugs Capomulin and Ramicane had the lowest final tumor volume between the four treatment groups. 
3) Looking at mouse s185 across timepoints, Capomulin decreased tumor volume from 45 mm3 to 25.47 mm3--a  43% reduction.
4) Looking at the Capomulin group, there was a strong statiscially signficant postive correlation of mouse weight and avg tumor volume (r = 0.84, p-value = 1.03e0-7).
5) Calcuation of the linear regression model shows a coefficient of determination (r-squared) of 0.709,  This represents the percent of the data that is the closest to the line of best fit.  Since r = 0.84 and r 2 = 0.709, means 70.9% of the total variation in  averge tumor volume can be explained by the linear relationship between mouse weight and average tumor volume (as described by the regression equation).  The other 29.1% of the total variation in average tumor volume remains unexplained.

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin

# Start by getting the last (greatest) timepoint for each mouse
max_tumor = clean_mouse_data.groupby([ 'Mouse ID'])['Timepoint'].max()
max_tumor = max_tumor.reset_index()


# Merge this group df with the original dataframe to get the tumor volume at the last timepoint
merged_data = max_tumor.merge(clean_mouse_data, on=['Mouse ID','Timepoint'], how='left')
merged_data.head(25)

```

```python
# Put treatments into a list for for loop (and later for plot labels)
RX_labels = ['Capomulin', 'Ceftamin','Infubinol', 'Ramicane']


# Create empty list to fill with tumor vol data (for plotting)
tumor_vol_data = []

# Calculate the IQR and quantitatively determine if there are any potential outliers. 
# Locate the rows which contain mice on each drug and get the tumor volumes


for drug in RX_labels:
    final_tumor_vol = merged_data.loc[merged_data['Drug Regimen'] == drug, "Tumor Volume (mm3)"]
    tumor_vol_data.append(final_tumor_vol)
    quartiles = final_tumor_vol.quantile([.25,.5,.75])
    lowerq = quartiles[.25]
    upperq = quartiles[0.75]
    iqr = upperq - lowerq
    lower_bound = lowerq - (1.5*iqr)
    upper_bound = upperq + (1.5*iqr)
    outliers = final_tumor_vol.loc[(final_tumor_vol < lower_bound) | (final_tumor_vol > upper_bound)]
    print(outliers)
   


    



```

```python
# Generate a box plot of the final tumor volume of each mouse across four regimens of interest
 
fig7, ax7 = plt.subplots()
ax7.set_title('Box Plot for 4 Drug Regimens Effects on Tumor Volume')
plt.ylabel('Final Tumor Volume (mm3)')
plt.xticks([1, 2, 3, 4], RX_labels) 
ax7.boxplot(tumor_vol_data)
plt.show()
    
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and linear regression model 
# for mouse weight and average tumor volume for the Capomulin regimen

correlation = st.pearsonr(weight, avg_tv)

print(f"The correlation coefficient (r) is {round(correlation[0],2)}")
print(f"The p-value is {correlation[1]}")
```

```python
x_values = weight
y_values = avg_tv

# Run the regression and store each of the returned values
(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)

# Calculate the predictions for each value of x (Mouse Weight)
regress_values = x_values * slope + intercept

# Create the linear equation
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))

print(f"The line of best fit: {line_eq}")
```

```python
plt.figure(figsize=(10,8))
# Plot observations
plt.scatter(x_values, y_values, alpha=0.8)

# Plot the regression line
plt.plot(x_values, regress_values, "red")

# Add the linear equation to the plot
plt.annotate(line_eq, (15, 44), fontsize=15, color="red")
plt.annotate(f"R2 = {round(rvalue**2,3)}", (15, 43), fontsize=15, color="red")

# Add labels
plt.xlabel('Mouse Weight (g)')
plt.ylabel('Average Tumor Volume (mm3)')

# calculate r2
print(f"The r-squared is: {rvalue**2}")

plt.show()
```
